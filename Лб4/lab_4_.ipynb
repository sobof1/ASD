# –õ–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–∞ —Ä–æ–±–æ—Ç–∞ ‚Ññ4. –†–µ–∫—É—Ä—Å—ñ—è. –°—Ç—Ä–∞—Ç–µ–≥—ñ—è ¬´—Ä–æ–∑–¥—ñ–ª—è–π —ñ –≤–æ–ª–æ–¥–∞—Ä—é–π¬ª

–í–∏–∫–æ–Ω–∞–≤ - –û–≥–æ–Ω–æ–≤—Å—å–∫–∏–π –û–ª–µ–∫—Å–∞–Ω–¥—Ä

**–ú–µ—Ç–∞**:  –∑–∞—Å–≤–æ—ó—Ç–∏ —Ç–µ—Ö–Ω–æ–ª–æ–≥—ñ—é —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–∏—Ö –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ –∑–∞—Å–æ–±–∞–º–∏ Python —ñ –æ—Ü—ñ–Ω—é–≤–∞–Ω–Ω—è —ó—Ö —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –≥–æ–ª–æ–≤–Ω–æ—ó —Ç–µ–æ—Ä–µ–º–∏ —Ä–µ–∫—É—Ä—Å—ñ—ó.


## –†–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ñ–∞–∫—Ç–æ—Ä—ñ–∞–ª—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ü–∏–∫–ª—É for. –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É.

```python
def find_factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –¥–∞–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É ‚Äî O(n), –æ—Å–∫—ñ–ª—å–∫–∏ —Ü–∏–∫–ª –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è n —Ä–∞–∑—ñ–≤. 

–£ –≤–∏–≥–ª—è–¥—ñ —Ñ–æ—Ä–º—É–ª–∏: $T(n) = c2 + c3n + c4(n - 1) + c5 = (c3 + c4) \cdot n + (c2 - c4 + c5)$, –¥–µ c ‚Äî –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ —Ä—è–¥–∫–∞ –∫–æ–¥—É.

## –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ñ–∞–∫—Ç–æ—Ä—ñ–∞–ª—É.

```python
def fac(n):
    # –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Ñ–∞–∫—Ç–æ—Ä—ñ–∞–ª–∞ —á–µ—Ä–µ–∑ —Ä–µ–∫—É—Ä—Å—ñ—é
    if n == 0:
        return 1
    return fac(n-1) * n

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
print(fac(5))
```

$T(n) = C_1 (n+1) + C_2 +C_3 n = (C_1 + C_3)n +(C_1 +C_2 )= an+b=n$

T(n) = O(n)

## –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è ùëõ-–≥–æ —á–∏—Å–ª–∞ –§—ñ–±–æ–Ω–∞—á—á—ñ.

```python
def fibonacci(n):

#–†–µ–∫—É—Ä—Å–∏–≤–Ω–µ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è n-–≥–æ —á–∏—Å–ª–∞ –§—ñ–±–æ–Ω–∞—á—á—ñ

    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(30))
```

$T(n) = C_1 \cdot T(n)+C_2+C_3\cdot T(n) + C_4 +C_5 \cdot T(n) = (C_1 +C_3 +C_5) \cdot T(n) + (C_2+C_4) = a \cdot T(n) + b$

$T(n) = O(2^n)$

## –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –∑–ª–∏—Ç—Ç—è–º, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≥–æ–ª–æ–≤–Ω—É —Ç–µ–æ—Ä–µ–º—É —Ä–µ–∫—É—Ä—Å—ñ—ó. 

```python
def merge(left, right):
    # –ó–ª–∏–≤–∞—î –¥–≤–∞ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏—Ö –º–∞—Å–∏–≤–∏ left —ñ right —É –æ–¥–∏–Ω
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # –î–æ–¥–∞—î–º–æ –∑–∞–ª–∏—à–∫–æ–≤—ñ –µ–ª–µ–º–µ–Ω—Ç–∏
    result += left[i:]
    result += right[j:]
    
    return result

def mergesort(lst):
    # –†–µ–∫—É—Ä—Å–∏–≤–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó
    if len(lst) < 2:
        return lst
    middle = len(lst) // 2
    left = mergesort(lst[:middle])
    right = mergesort(lst[middle:])
    return merge(left, right)

# –ü—Ä–∏–∫–ª–∞–¥ –º–∞—Å–∏–≤—É
a = [6, 5, 3, 1, 8, 7, 2, 4]

# –í–∏–∫–ª–∏–∫ —Ñ—É–Ω–∫—Ü—ñ–π
sorted_a = mergesort(a)
print(sorted_a)  # –í–∏–≤–µ–¥–µ: [1, 2, 3, 4, 5, 6, 7, 8]
```

- a = 2 (–¥—ñ–ª–µ–º–æ –Ω–∞ 2 –ø—ñ–¥–∑–∞–¥–∞—á—ñ)
- b = 2 (—Ä–æ–∑–º—ñ—Ä –∑–º–µ–Ω—à—É—î—Ç—å—Å—è –≤–¥–≤—ñ—á—ñ)
- d = 1 (–∑–ª–∏—Ç—Ç—è –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∑–∞ O(n))

$log_b  a = log_2 2 = 1$

–æ—Å–∫—ñ–ª—å–∫–∏ $d = log_b a$, —Ç–æ $t(n) = O(n log n)$
